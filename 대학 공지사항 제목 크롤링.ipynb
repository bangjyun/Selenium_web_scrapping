## qna 크롤링

```python
# 로그인 웹에 접속
from selenium import webdriver
import time
browser = webdriver.Chrome("~/chromedriver.exe")
browser.get("https://www.ajou.ac.kr/kr/bachelor/bamedia-qna.do?auty=2")  
```

```python
# 자동 로그인
browser.find_element_by_id("userId").send_keys("IDD")
browser.find_element_by_id("password").send_keys("PWW!")
browser.find_element_by_id("loginSubmit").click()
```

#### css selector 이용
locates = "td.b-td-left > div > a"
elements = browser.find_elements_by_css_selector(locates)   
for element in elements:
    print(element.text)

```python
# selenium 함수 이용
eles = browser.find_elements_by_class_name("b-title-box")
for e in eles:
    print(e.text)
```

### 공통 attribute 가 애매할 때 

```python
for i in range(101,500):
    url = "https://www.ajou.ac.kr/kr/bachelor/bamedia-qna.do?auty=2&auty=2"+"?mode=list&&articleLimit=10&article.offset={}".format(i*10)
    browser.get(url)
    locates = "td.b-td-left > div > a"
    elements = browser.find_elements_by_css_selector(locates)
    for element in elements:
        print(element.text)
```
